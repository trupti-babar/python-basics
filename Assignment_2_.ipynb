### Question 3: An e-commerce platform has customer IDs from two campaigns:
campaign1 = {101, 102, 103, 104, 105}

campaign2 = {104, 105, 106, 107}
1. Find unique customers across both campaigns.
2. Find customers common in both.
3. Find customers exclusive to campaign1.
4. Explain how sets help in removing duplicates in real DS projects.

```python
campaign1 = {101, 102, 103, 104, 105}

campaign2 = {104, 105, 106, 107}
```

```python
# 1.Unique customers across both campaigns
unique_customers = campaign1.union(campaign2)

print("Unique customers:", unique_customers)
```

```python
# 2.Common customers in both campaigns
common_customers = campaign1.intersection(campaign2)

print("Common customers:", common_customers)
```

```python
# 3.Customers exclusive to campaign1
exclusive_customers = campaign1.difference(campaign2)

print("Exclusive customers in campaign1:", exclusive_customers)
```

4.Sets are very important in real Data Science (DS) projects because they automatically remove duplicate values and keep only unique data.

Duplicates are common in actual datasets, for instance:

Multiple registrations by one client,
Multiple occurrences of the same transaction,
Duplicated IDs in large databases.

This is why we use sets to easily clean such data.

### Question 5: Given:
sales = [12000, 15000, 11000, 18000, 17000]

Using a loop:
1. Calculate total revenue.
2. Count how many months had sales above 15000.
3. Find maximum sales without using built-in max(

```python
sales = [12000, 15000, 11000, 18000, 17000]
```

```python

# 1.Calculate total revenue
total_revenue = 0

for amount in sales:
    total_revenue += amount

print("Total Revenue:", total_revenue)
```

```python
# 2. Count months with sales above 15000
count = 0

for amount in sales:
    if amount > 15000:
        count += 1

print("Months with sales above 15000:", count)
```

```python
# 3. Find maximum sales without max()

maximum = sales[0]

for amount in sales:
    if amount > maximum:
        maximum = amount

print("Maximum Sales:", maximum)
```

### Question 6: You are analyzing product ratings:
ratings = [

[5, 4, 3],

[4, 4, 5],

[3, 5, 4]

]
1. Compute average rating.
2. Print all ratings greater than 4.
3. Explain how nested loops are used in matrix-based ML computations.


```python
ratings = [
    [5, 4, 3],
    [4, 4, 5],
    [3, 5, 4]
]
```

```python

# 1.Compute average rating
total = 0
count = 0

for row in ratings:
    for rating in row:
        total += rating
        count += 1

average = total / count

print("Average Rating:", average)
```

```python
# 2.Print ratings greater than 4
print("Ratings greater than 4:")

for row in ratings:
    for rating in row:
        if rating > 4:
            print(rating)
```

3.Nested loops are used in matrix-based ML computations because Machine Learning data is often stored in rows and columns called matrices.
The outer loop iterates on the rows, and the inner loop iterates on the values within each row.

They are used in:

matrix operations.
image processing.
neural networks and
big data.

### Question 8: Given
prices = [500, 1500, 2500, 800, 1200]
1. Use map() + lambda to add 18% GST to all prices.
2. Use filter() to select prices above 1000.
3. Explain where such transformations are used in real-world pipelines

```python
prices = [500, 1500, 2500, 800, 1200]
```

```python
# 1.Add 18% GST using map() and lambda
gst_prices = list(map(lambda x: x + (x * 18 / 100), prices))

print("Prices after GST:", gst_prices)
```

```python
# 2.Filter prices above 1000
filtered_prices = list(filter(lambda x: x > 1000, prices))

print("Prices above 1000:", filtered_prices)
```

3.These kinds of transformations are utilized in practice for preprocessing and cleansing purposes.In `map()`, we modify or change data values, whereas in `filter()`, we choose the data that we need.

They enable us to clean up our data for further analysis and modeling in Machine Learning.

### Question 9:Write a generator that yields squares of numbers from 1 to 1,000,000.
Explain difference between:

● List comprehension

● Generator expression

Why are generators preferred in big data scenarios?

```python
# 1.generator function
def square_generator():

    for i in range(1, 1000001):
        yield i ** 2

# create generator object
squares = square_generator()

# print first 5 square values
for _ in range(5):
    print(next(squares))
```

2.**List Comprehension**

* Stores all values in memory at once

* Uses more memory

* Faster for small datasets

* Example:

   [x**2 for x in range(10)]

**Generator Expression**

* Generates values one by one

* Uses less memory

* Better for large datasets

* Example:

   (x**2 for x in range(10))

3.Generators are preferred in big data situations because they produce values one at a time instead of loading all the data into memory at once.

This approach saves memory and improves performance when dealing with very large datasets.

Generators are useful in:

* big data processing
* data streaming
* machine learning pipelines
* handling large files


### Question 10: Design a class:
class Product:

With:

● Attributes: name, price, quantity

● Method: calculate_revenue()

Create 3 objects and print their revenue.

```python
class Product:

    # constructor
    def __init__(self, name, price, quantity):

        self.name = name
        self.price = price
        self.quantity = quantity

    # method to calculate revenue
    def calculate_revenue(self):
          return self.price * self.quantity

# create objects
product1 = Product("Laptop", 50000, 2)
product2 = Product("Mobile", 20000, 3)
product3 = Product("Headphones", 1500, 5)

# print revenues
print(product1.name, "Revenue:", product1.calculate_revenue())
print(product2.name, "Revenue:", product2.calculate_revenue())
print(product3.name, "Revenue:", product3.calculate_revenue())
```
